## Dataset Loader

This notebook demonstrates how to use the AIICE loader to:

- Inspect dataset metadata
- Download dataset files locally
- Load time-series data as NumPy arrays or PyTorch tensors
- Visualize dataset samples

```python
from pprint import pprint
from datetime import date
import matplotlib.pyplot as plt
```

```python
from aiice.loader import Loader

loader = Loader()
```

### Overview

The loader exposes basic dataset information with some stats values:

- `shape` — spatial resolution of each sample
- `dataset_start` — first available date
- `dataset_end` — last available date
- `seas` - available seas

```python
print(loader.shape, loader.dataset_start, loader.dataset_end)
pprint(loader.seas)
```

```python
info = loader.info(per_year=True)

pprint(info)
```

## Download / Get files

```python
start_date = date(1999, 11, 1)
end_date = date(2000, 2, 1)
```

You can fetch files to a local disk as a raw numpy matrices. Let's download dates from 1999.11.01 to 2000.02.01 per 3 days:

```python
file_paths = loader.download(local_dir="../../../.tmp/dowload_example", start=start_date, end=end_date, step=3)
```

```python
file_paths
```

Or you can get files in your RAM and convert it to the numpy matrices or pytorch tensors. Let's get dates from 1999.11.01 to 2000.02.01 per 3 days:

```python
data = loader.get(start=start_date, end=end_date, step=3, tensor_out=True)
```

`start` and `end` params are optional and equals to `dataset_start` and `dataset_end` values by default. 
Moreover date args can be initialized as `str`.
Let's try to visualize one day of one sea:

```python
one_day = loader.get(start="2021-01-01", end="2021-01-01", sea="Bering Sea")
```

Matrices have only ice density values as `float` from 0 to 1:

```python
plt.figure(figsize=(8, 6))
plt.imshow(one_day[0])
plt.title(f'Bering Sea')
plt.show()
```
